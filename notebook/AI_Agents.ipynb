```python
# ==================================================
# 📌 Importing Required Libraries
# ==================================================
import os
import openai
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

# ==================================================
# 🔑 OpenAI API Key Setup
# ==================================================
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if not OPENAI_API_KEY:
    OPENAI_API_KEY = input("Enter your OpenAI API Key: ").strip()
    os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

# ==================================================
# 🏋️‍♂️ User Details Input
# ==================================================
print("\n🤖 AI Fitness Coach: Hi! Before we start, can I get some details?")

user_details = {
    "age": input("Enter Age: ").strip(),
    "weight": input("Enter Weight (kg): ").strip(),
    "height": input("Enter Height (cm): ").strip(),
    "gender": input("Enter Gender (M/F): ").strip(),
    "activity_level": input("Enter Activity Level (Sedentary, Moderate, Active): ").strip(),
    "diet_style": input("Enter Diet Style (Veg, Vegan, Non-Veg, Keto, etc.): ").strip(),
    "body_type": input("Enter Your Goal Body Type (Lean, Muscular, Athletic, etc.): ").strip()
}

# ==================================================
# ✨ Prompt Template for AI Responses
# ==================================================
prompt_template = """
You are an AI Health & Fitness Coach. Analyze user requests and respond accordingly.

User Details:
- Age: {age}
- Weight: {weight}kg
- Height: {height}cm
- Gender: {gender}
- Activity Level: {activity_level}
- Diet Style: {diet_style}
- Body Type Goal: {body_type}

User Request: {user_input}

Assistant:
"""

# ==================================================
# 🚀 Initialize OpenAI LLM for Conversations
# ==================================================
llm = ChatOpenAI(model_name="gpt-4", temperature=0.7)

# ==================================================
# 🗨️ Interactive Chat Function
# ==================================================

from langchain.schema import HumanMessage # Import the HumanMessage class
def chatbot():
    print("\n🤖 AI Fitness Coach: Hello! I'm your personal AI health & fitness assistant. Let's optimize your health! 🏋️‍♂️🥗")

    while True:
        user_input = input("\nYou: ").strip().lower()

        # Exit command
        if user_input in ["exit", "quit", "bye"]:
            print("\n🤖 AI Fitness Coach: Goodbye! Stay fit and healthy! 💪")
            break

        formatted_input = user_input
        if "calorie" in user_input:
            formatted_input = (
                f"Assume you are a world class dietician working in a top hospital. Estimate the daily calorie needs for a {user_details['age']}-year-old, "
                f"{user_details['weight']}kg, {user_details['height']}cm {user_details['gender']} "
                f"with {user_details['activity_level']} activity level."
                f"Provide the answer in 5 words strictly."
            )
        elif "meal plan" in user_input:
            formatted_input = (
                f"Create a structured meal plan for a {user_details['diet_style']} diet "
                f"that maintains a balanced intake of proteins, carbs, fats, and vitamins."
                f"Finally, Provide an excel like table to summarize the results by seperating each column by '|'. Structure the column spacing equally between all columns strictly enabling a very good view for the user. Do not show any other calculations/ steps other than tables."
            )
        elif "workout" in user_input:
            formatted_input = (
                f"Suggest a workout plan for someone aiming for a {user_details['body_type']} body type, "
                f"considering their calorie intake and fitness goals."
                f"give me a workout plan with nuber of hours, what to workout, if I want a {user_details['body_type']} body type."
                f"Finally, Provide an excel like table to summarize the results by seperating each column by |. Structure the column spacing equally between all columns strictly enabling a very good view for the user. Do not show any other calculations/ steps other than tables."
            )

        # Format the prompt with user input
        formatted_prompt = prompt_template.format(
            age=user_details["age"],
            weight=user_details["weight"],
            height=user_details["height"],
            gender=user_details["gender"],
            activity_level=user_details["activity_level"],
            diet_style=user_details["diet_style"],
            body_type=user_details["body_type"],
            user_input=user_input
        )

        # Create a HumanMessage object
        message = HumanMessage(content=formatted_prompt)

        # Get response from LLM
        response = llm([message]) # Pass a list containing the HumanMessage

        # Print AI response
        print(f"\n🤖 AI Fitness Coach: {response.content}") # Access content directly for chat models
chatbot()
```

```python
import openai
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain.chat_models import ChatOpenAI

# OpenAI API Key
openai.api_key = "your-openai-api-key"

# Initialize Memory
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# Prompt Template -
prompt_template = PromptTemplate(
    input_variables=["chat_history", "combined_input"],
    template="""
    You are an AI Job Recommender. Based on the user's resume and the job description provided,
    suggest the best fit job roles, and tell them why they are a good match. Also, Provide Job postings where the user can apply to. Make sure this is in a structured table with
    Company, Job Role, pay and link to the job posting as columns seperated by | and all columns have equal spacing and looks really good to the user. Give the user real job postings where they can apply. Do not show any other calculations/ steps other than tables.


    Chat History:
    {chat_history}

    Combined Input:
    {combined_input}

    Provide your recommendation:
    """
)

# Define prompt template for resume improvement
resume_improvement_prompt = PromptTemplate(
    input_variables=["user_resume", "job_description"],
    template="""
    You are an AI that helps tailor resumes for specific job roles. Analyze the user's resume and the
    provided job description, and suggest improvements to the resume. Your goal is to optimize the resume
    for the given role by adding relevant skills, experience, and keywords.

    User Resume:
    {user_resume}

    Job Description:
    {job_description}

    Suggest improvements to the resume:
    """
)

# Initialize LangChain with OpenAI LLM
llm = ChatOpenAI(model_name="gpt-4o", temperature=0.7)

# chatbot chain with memory and prompt template
chatbot_chain = LLMChain(
    llm=llm,
    prompt=prompt_template,
    memory=memory
)

# resume improvement chain
resume_improvement_chain = LLMChain(
    llm=llm,
    prompt=resume_improvement_prompt
)

def job_recommender_chatbot():
    print("🤖 Job Recommender Chatbot: Hello! Please enter your resume below for job recommendations.")

    user_profile = input("You: Please paste your resume here: ")
    job_description = input("You: Please enter the job description you're interested in: ")

    # Combine user profile (resume) and job description into one string for recommendation
    combined_input = f"Resume: {user_profile}\n\nJob Description: {job_description}"

    # job recommendation
    job_recommendation = chatbot_chain.run(combined_input=combined_input, chat_history="")
    print(f"\n🤖 Job Recommender Chatbot: {job_recommendation}")

    # resume modifications
    improved_resume = resume_improvement_chain.run(user_resume=user_profile, job_description=job_description)

    print(f"\n🤖 Resume Improvement Tool: Here's how you can improve your resume for the role:")
    print(improved_resume)

# Run the chatbot
job_recommender_chatbot()
```
